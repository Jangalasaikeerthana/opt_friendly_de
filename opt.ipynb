```python
import re

def cleaning_name(name):
    name = re.sub(r'\b(llc|ltd|inc|corp|co|lp|llp|pvt|limited|corporation|)\b','',name,flags=re.IGNORECASE).lower()

    name = re.sub(r'[^a-z0-9]', '', name) 
    return name

def get_data_jobs(slug):
    slug = cleaning_name(slug)
    url = f"https://boards-api.greenhouse.io/v1/boards/{slug}/jobs"
    res= requests.get(url)


    if res.status_code!=200:
        return []
    jobs_list = res.json()['jobs']
    matches=[]
    for i in jobs_list:
        if 'data' in i['title'].lower():
            matches.append({'company':slug,
                            'title':i['title'],
                            'location':i['location']['name'],
                            'url': i['absolute_url']

            
                            })
    return matches
    
get_data_jobs('Akuna Capital LLC')
```

```python
slugs= dfo.iloc[:, 0].tolist()
big=[]
for s in slugs:

    results= get_data_jobs(s)
    big.extend(results)
print(big)
```

```python
found = 0
not_found = 0

for s in slugs:
    slug = cleaning_name(s)
    url = f"https://boards-api.greenhouse.io/v1/boards/{slug}/jobs"
    res = requests.get(url)
    if res.status_code == 200:
        found += 1
    else:
        not_found += 1

print('On Greenhouse:', found)
print('Not found:', not_found)
```
